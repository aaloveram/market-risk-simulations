# FE Micro-Simulation

```python
# libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Concept: Daily returns of an asset:
In Financial Engineering, we often model returns, not prices.

## Assumptions:
- Average Daily Return = 0.05%
- Daily Volatility = 1%
- 252 Trading Days (1 Year)

## Purpose of simulation
Pretend you are the stock market. Most days you grow a tiny bit. Some days you go up or down more. Do this for one year. Behave the same every time I ask.

```python
# code
np.random.seed(42) # seed starts randomness from the same starting point

# this variable answers how much asset went up or down each day
daily_returns = np.random.normal( # provides numbers that behave like the real world
    loc=0.0005, # mean return and center of the bell curve for the normal distribution
    scale=0.01, # volatility / risk
    size=252 # provides 1 year of daily behavior
)
```

### Explanation for code above
- Each number = one day's return
- This signals a stochastic process as well as market uncertainty simulation

```python
# code
daily_returns[:10]
```

### Explanation of code above
- Negative does not signal a bad model, it's just how it works in real life
- Markets go up and down all the time

```python
# bell curve code simulator
plt.figure()
plt.hist(daily_returns, bins=30)
plt.title("Distribution of Daily Returns")
plt.xlabel("Daily Return")
plt.ylabel("Number of Days")
plt.show()
```

### Explanation for code above
- Most days are near 0, which signal small ups and downs
- Right side highlight positive returns; left side portrays the losses
- Average is positive

### Reflection over mini-simulation
Even with a positive expected return, some individual days can still be negative. Volatility causes dispersion around the mean, which represents risk.
